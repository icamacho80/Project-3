```
filteredSalesUSAdate_df=filteredSalesUSA_df
filteredSalesUSAdate_df['Date'] = pd.to_datetime(filteredSalesUSAdate_df['Date'], format='%m/%d/%Y')
```

```
filteredSalesUSAdate_df.dtypes
```

```
filteredSalesUSAdate_df['YearMonth'] = filteredSalesUSAdate_df['Date'].dt.to_period('M')  
total_sales_per_month = filteredSalesUSAdate_df.groupby('YearMonth')['Total_Amount'].sum()  
```

```
print(total_sales_per_month)
```

```
plt.figure(figsize=(10, 6))  # Set the size of the plot
ax= total_sales_per_month.plot(kind='bar')

# Step 2: Customize the plot
plt.title('Seasonality Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales (in $M)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

for i, value in enumerate(total_sales_per_month):
    ax.text(i, value + 0.005 * value, f'{value/1000000:.2f}', ha='center', va='bottom', fontsize=10)

average_sales = total_sales_per_month.mean()
plt.axhline(y=average_sales, color='r', linestyle='--', label=f'Average Sales: {average_sales/1000000:.1f}M')
plt.legend()

formatter = FuncFormatter(lambda x, _: f'{x/1000000:.0f}M')  # Divide by 1 million and add 'M'
ax.yaxis.set_major_formatter(formatter)

plt.ylim(bottom=8000000)

# Step 3: Show the plot


plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```
